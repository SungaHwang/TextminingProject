### 직무_lv1 unique 값 찾기

```python
# 직무_lv1 열의 value_counts 구하기
job_lv1 = gongo['직무_lv1'].value_counts().index
job_lv1
```

```python
# 직무_lv1이 하나의 값만 갖는 경우 리스트에 추가
job_lv1_list = []
for job in job_lv1:
  if len(job)==1:
    job_lv1_list.append(job[0])
job_lv1_list
```

### 직무_lv2 unique 값 찾기

```python
# 직무_lv2 열의 value_counts 구하기
job_lv2 = gongo['직무_lv2'].value_counts().index
job_lv2
```

```python
# 직무_lv1이 하나의 값만 갖는 경우 리스트에 추가
job_lv2_list = []
for job in job_lv2:
  if len(job)==1:
    job_lv2_list.append(job[0])
job_lv2_list
```

### 직무 lv1 데이터프레임 생성

```python
# 직무 lv1 데이터프레임 생성
job_lv1_df = pd.DataFrame(columns = job_lv1_list)
job_lv1_df
```

```python
# 직무 lv1 데이터프레임에 각 데이터별 해당되는 직무 컬럼이 있으면 1값 부여
for i, job in enumerate(gongo['직무_lv1']):
  for category_job in job_lv1_list:
    if category_job in job:
      job_lv1_df.loc[i,category_job] = 1
job_lv1_df.fillna(0, inplace=True)
```

보통 글이 잘못 들어간 컬럼: 기업 소개, 주요 업무, 자격 요건, 우대 사항

```python
# 기업 소개 또는 우대 사항 컬럼이 nan인 데이터: 759
gongo_nan = gongo[gongo['기업 소개'].isna() | gongo['우대 사항'].isna()]
gongo_nan.reset_index(inplace=True)
gongo_nan = gongo_nan[['index','주요 업무','자격 요건','우대 사항']]
```

```python
# check
gongo_nan
```

```python
# 자격요건+우대사항 하나의 컬럼으로 합치기
gongo_nan['자격요건 및 우대사항'] = gongo_nan['자격 요건']+gongo_nan['우대 사항']
# 자격요건, 우대사항 각 컬럼 삭제
gongo_nan.drop(['자격 요건','우대 사항'],axis=1, inplace=True)
gongo_nan # check
```

```python
# 사전
position = '담당업무|담당 업무|주요업무|주요 업무|직무설명|직무 설명|어떤 일을 하나요?|직무소개|직무 소개' # 주요 업무
qualification = '지원자격|지원 자격|필수조건|필수 조건|공통자격|공통 자격|자격요건|자격 요건' # 자격 요건
prefer = '우대사항|우대사항' # 우대 사항

position_list = ['담당업무','담당 업무','주요업무','주요 업무','직무설명','직무 설명','어떤 일을 하나요?','직무소개','직무 소개']
qualification_list = ['지원자격','지원 자격','필수조건','필수 조건','공통자격','공통 자격','자격요건','자격 요건']
prefer_list = ['우대사항','우대 사항']

# company = '회사설명|회사 설명|기업 소개|기업소개|팀 소개|팀소개' # 기업 소개
# hiring_process = '합류과정|합류 과정|채용절차|채용 절차' # 채용절차
```

```python
# 주요 업무 컬럼
pos1 = gongo_nan['주요 업무'].str.contains(position)
qual1 = gongo_nan['주요 업무'].str.contains(qualification)
pref1 = gongo_nan['주요 업무'].str.contains(prefer)

pos1_index = pos1[pos1 == True].index
qual1_index = pos1[pos1 == True].index
pref1_index = pos1[pos1 == True].index


# 자격요건 및 우대사항 컬럼
pos2 = gongo_nan['자격요건 및 우대사항'].str.contains(position)
qual2 = gongo_nan['자격요건 및 우대사항'].str.contains(qualification)
pref2 = gongo_nan['자격요건 및 우대사항'].str.contains(prefer)

pos2_index = pos2[pos2 == True].index
qual2_index = pos2[pos2 == True].index
pref2_index = pos2[pos2 == True].index
```

```python
# 주요 업무 컬럼
# 주요업무 셀에 position이 들어간 경우: 그 앞은 내용 자르기
for i in pos1_index:
  for pos_word in position_list:
    pos_word_index = gongo_nan['주요 업무'][i].find(pos_word)
    if pos_word_index!=-1: # 주요업무에 다른 내용이 있는 경우
      gongo_nan['주요 업무'][i] = gongo_nan['주요 업무'][i][pos_word_index:]

# 주요업무 셀에 qualification이 들어간 경우: 그 뒤 내용 자르기
for i in qual1_index:
  for qual_word in qualification_list:
    qual_word_index = gongo_nan['주요 업무'][i].find(qual_word)
    if qual_word_index!=-1: # 주요업무에 다른 내용(qualication)이 있는 경우
      # print(gongo_nan['주요 업무'][i])
      # print('substring')
      # print(gongo_nan['주요 업무'][i][:qual_word_index])
      # print('----------')
      gongo_nan['주요 업무'][i] = gongo_nan['주요 업무'][i][:qual_word_index]

# 주요 업무 셀에 prefer이 들어간 경우: 그 뒤 내용 자르기
for i in pref1_index:
  for prefer_word in prefer_list:
    prefer_word_index = gongo_nan['주요 업무'][i].find(prefer_word)
    if prefer_word_index!=-1: # 다른 내용이 있는 경우
      gongo_nan['주요 업무'][i] = gongo_nan['주요 업무'][i][:prefer_word_index]
```

```python
# 자격요건 및 우대사항 컬럼
# 자격요건 및 우대사항 셀에 qualification이 들어간 경우: 그 앞 내용 자르기
for i in qual2_index:
  for qual_word in qualification_list:
    qual_word_index = gongo_nan['자격요건 및 우대사항'][i].find(qual_word)
    if qual_word_index!=-1: # 자격요건 및 우대사항 다른 내용(qualication)이 있는 경우
      # print(gongo_nan['주요 업무'][i])
      # print('substring')
      # print(gongo_nan['주요 업무'][i][:qual_word_index])
      # print('----------')
      gongo_nan['자격요건 및 우대사항'][i] = gongo_nan['자격요건 및 우대사항'][i][qual_word_index:]
```
