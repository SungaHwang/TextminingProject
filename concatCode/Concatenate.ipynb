```python
# concatenate dataframes
filenames = os.listdir(folder_path)

review = pd.DataFrame()
recruitment = pd.DataFrame()

import pandas as pd

for filename in filenames[1:10]:
  jobname, datatype = filename.strip('.csv').split('_')
  df = pd.read_csv(folder_path+'/'+filename)

  # 데이터프레임에 직종 컬럼 추가
  df['직종'] = jobname

  if datatype == '공고':
    recruitment = recruitment.append(df)
  elif datatype == '리뷰':
    review = review.append(df)
```
