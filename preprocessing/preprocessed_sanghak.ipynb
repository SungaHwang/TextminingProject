```python
import pandas as pd
import numpy as np
import re
import string
```

```python
df = pd.read_csv('/content/drive/MyDrive/[TM] Term Project/result/preprocessed_gongo_final.csv')
df
```

```python
def rmEmoji(inputData):
    return inputData.encode('utf-8', 'ignore').decode('utf-8')
```

```python
df['자격요건 및 우대사항'][:3]
```

```python
for i, strrrr in enumerate(df['자격요건 및 우대사항']):
  if type(strrrr)!=str:
    continue
  else:
    rmEmoji(strrrr)
    df['자격요건 및 우대사항'][i] = strrrr.translate(str.maketrans('', '', string.punctuation)).replace('\r', "").replace("\n", "").replace('ㆍ', "").replace('•', "").replace('\t', "")

for i, strrrr in enumerate(df['주요 업무']):
  if type(strrrr)!=str:
    continue
  else:
    rmEmoji(strrrr)
    df['주요 업무'][i] = strrrr.translate(str.maketrans('', '', string.punctuation)).replace('\r', "").replace("\n", "").replace('ㆍ', "").replace('•', "").replace('\t', "").replace('"',"").replace("'","")


for i, strrrr in enumerate(df['복리후생']):
  if type(strrrr)!=str:
    continue
  else:
    rmEmoji(strrrr)
    df['복리후생'][i] = strrrr.translate(str.maketrans('', '', string.punctuation)).replace('\r', "").replace("\n", "").replace('ㆍ', "").replace('•', "").replace('\t', "").replace('"',"").replace("'","")

for i, strrrr in enumerate(df['채용 절차']):
  if type(strrrr)!=str:
    continue
  else:
    rmEmoji(strrrr)
    df['채용 절차'][i] = strrrr.translate(str.maketrans('', '', string.punctuation)).replace('\r', "").replace("\n", "").replace('ㆍ', "").replace('•', "").replace('\t', "").replace('"',"").replace("'","")

for i, strrrr in enumerate(df['기업 소개']):
  if type(strrrr)!=str:
    continue
  else:
    rmEmoji(strrrr)
    df['기업 소개'][i] = strrrr.translate(str.maketrans('', '', string.punctuation)).replace('\r', "").replace("\n", "").replace('ㆍ', "").replace('•', "").replace('\t', "").replace('"',"").replace("'","")
```
