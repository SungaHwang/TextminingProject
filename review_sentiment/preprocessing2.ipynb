```python
import pandas as pd
import numpy as np
```

```python
review_df = pd.read_csv('C:/TextminingProject//concatdataset/review_df.csv')
print(len(review_df))
```

```python
review_df['직무'].unique()
```

```python
delete1 = review_df[review_df['직무'] == '\n          현직원\n      '].index
delete2 = review_df[review_df['직무'] == '\n          전직원\n      '].index
review_df = review_df.drop(delete1)
review_df = review_df.drop(delete2)
```

```python
review_df = review_df.dropna()
review_df['직무'].unique()
```

```python
review_df['장점'] = review_df['장점'].str.replace("\n","")
review_df['장점'] = review_df['장점'].str.replace("\r","")
review_df['장점'] = review_df['장점'].str.replace("\t","")
review_df['장점'] = review_df['장점'].str.replace(",","")
review_df['장점'] = review_df['장점'].str.replace(".","")
```

```python
review_df['단점'] = review_df['단점'].str.replace("\n","")
review_df['단점'] = review_df['단점'].str.replace("\r","")
review_df['단점'] = review_df['단점'].str.replace("\t","")
review_df['단점'] = review_df['단점'].str.replace(",","")
review_df['단점'] = review_df['단점'].str.replace(".","")
```

```python
review_df['요약'] = review_df['요약'].str.replace("\n","")
review_df['요약'] = review_df['요약'].str.replace("\r","")
review_df['요약'] = review_df['요약'].str.replace("\t","")
review_df['요약'] = review_df['요약'].str.replace(",","")
review_df['요약'] = review_df['요약'].str.replace(".","")
```

```python
review_df = review_df[['직무', '장점', '단점','별점']]
review_df
```

```python
review_df['장점']= review_df['장점'].str.replace(pat=r'[^\w]',repl=r' ',regex=True)
review_df['단점']= review_df['단점'].str.replace(pat=r'[^\w]',repl=r' ',regex=True)
```

```python
review_df_type = review_df.groupby('직무')['별점']
review_df_type.count()
```

```python
review_df['별점'].value_counts()
```

```python
review_df['리뷰'] = review_df['장점'].map(str) + ". " + review_df['단점'].map(str)
review_df
```

```python
review_df_po = review_df[['직무', '장점']]
review_df_po['라벨'] = 1
review_df_po.rename(columns = {'장점' : '장단점'}, inplace = True)
review_df_ne = review_df[['직무', '단점']]
review_df_ne['라벨'] = 0
review_df_ne.rename(columns = {'단점' : '장단점'}, inplace = True)
review_df_new = pd.concat([review_df_po, review_df_ne], ignore_index=True)
review_df_new
```

```python
selected = ['데이터','디자인']
review_df_test =review_df_new[review_df_new['직무'].isin(selected)].reset_index(drop=True)
review_df_test
```

```python
review_df_test.to_csv('review_df_test_po.txt', sep ='\t', index = False)
```

```python
review_df_train = review_df_new[~review_df_new['직무'].isin(selected)].reset_index(drop=True)
review_df_train = review_df_train[['장점','라벨']]
review_df_train
```

```python
review_df_train['별점'] = review_df_train['별점'].astype(int)
review_df_train['별점']
```

```python
review_df_train['별점'] = review_df_train['별점'] -1
review_df_train['별점']
```

```python
review_df_train.to_csv('review_df_train2.txt', sep ='\t', index = False)
```
